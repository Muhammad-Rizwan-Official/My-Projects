```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
alzi.replace({'Group':{'Nondemented':0,'Demented':1,'Converted':2}},inplace=True)
```

```python
alzi['M/F']=le.fit_transform(alzi['M/F'])
alzi
```

```python
alzi['MMSE']=alzi['MMSE'].fillna(0)
```

```python
alzi['SES']=alzi['SES'].fillna(0)
```

```python
x=alzi.drop(['Group','Hand','Subject ID','MRI ID'],axis=1)
x
```

```python
x.isnull().sum()
```

```python
y=alzi['Group']
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=1)
```

```python
x.shape,X_train.shape,X_test.shape
```

```python
tree=DecisionTreeClassifier()
```

```python
tree.fit(X_train,y_train)
X_train_pred=tree.predict(X_train)
score=accuracy_score(y_train,X_train_pred)
print('accuracy score of trained data:',score)

tree.fit(X_test,y_test)
X_test_pred=tree.predict(X_test)
score=accuracy_score(y_test,X_test_pred)
print('accuracy score of test data:',score)
```

```python
fore=RandomForestClassifier()
```

```python
fore.fit(X_train,y_train)
X_train_pred=fore.predict(X_train)
score=accuracy_score(y_train,X_train_pred)
print('accuracy score of trained data:',score)

fore.fit(X_test,y_test)
X_test_pred=fore.predict(X_test)
score=accuracy_score(y_test,X_test_pred)
print('accuracy score of test data:',score)
```

```python
knn=KNeighborsClassifier()
```

```python
knn.fit(X_train,y_train)
X_train_pred=knn.predict(X_train)
score=accuracy_score(y_train,X_train_pred)
print('accuracy score of trained data:',score)

knn.fit(X_test,y_test)
X_test_pred=knn.predict(X_test)
score=accuracy_score(y_test,X_test_pred)
print('accuracy score of test data:',score)
```
