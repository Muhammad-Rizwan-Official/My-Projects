```python
df.replace({'y':{'tumor':1,'Normal':0}},inplace=True)
df
```

```python
plt.figure(figsize=(5,5))
sns.countplot(x='Class',data=df)
```

```python
y=df['Class']
y
```

```python
x=df.drop(['Image','Class','Coarseness'],axis=1)
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=1)
```

```python
x.shape,X_train.shape,X_test.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
linear=LinearRegression()
```

```python
linear.fit(X_train,y_train)
X_train_pred=linear.predict(X_train)
score=metrics.r2_score(X_train_pred,y_train)
print('Accuracy of training data',score)

linear.fit(X_test,y_test)
X_test_pred=linear.predict(X_test)
score=metrics.r2_score(X_test_pred,y_test)
print('Accuracy of testing data',score)
```

```python
plt.scatter(y_train,X_train_pred)
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg=LogisticRegression()
```

```python
logreg.fit(X_train,y_train)
X_train_pred=logreg.predict(X_train)
score=metrics.r2_score(X_train_pred,y_train)
print('Accuracy of training data',score)

logreg.fit(X_test,y_test)
X_test_pred=logreg.predict(X_test)
score=metrics.r2_score(X_test_pred,y_test)
print('Accuracy of testing data',score)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree=DecisionTreeClassifier()
```

```python
tree.fit(X_train,y_train)
X_train_pred=tree.predict(X_train)
score=metrics.r2_score(X_train_pred,y_train)
print('Accuracy of training data',score)

tree.fit(X_test,y_test)
X_test_pred=tree.predict(X_test)
score=metrics.r2_score(X_test_pred,y_test)
print('Accuracy of testing data',score)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
forest=RandomForestClassifier()
```

```python
forest.fit(X_train,y_train)
X_train_pred=forest.predict(X_train)
score=metrics.r2_score(X_train_pred,y_train)
print('Accuracy of training data',score)

forest.fit(X_test,y_test)
X_test_pred=forest.predict(X_test)
score=metrics.r2_score(X_test_pred,y_test)
print('Accuracy of testing data',score)
```

```python
from sklearn import svm
```

```python
cls=svm.SVC(kernel='linear')
```
